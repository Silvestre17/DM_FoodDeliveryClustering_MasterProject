```python
# Check for missing values in the dataset (n = number of missing values & % = percentage of missing values)
NAs_df = pd.DataFrame({'n NAs': ABCDEats.isnull().sum(), 
                       '% NAs': round(ABCDEats.isnull().mean() * 100,2)})
NAs_df.index.name = 'Variables'

# Display the variables with missing values
NAs_df[NAs_df['n NAs'] > 0]
```

---

```python
# Filtrar os casos onde customer_age estÃ¡ missing e not missing
missing_customer_age = ABCDEats[ABCDEats['customer_age'].isnull()]
not_missing_customer_age = ABCDEats[ABCDEats['customer_age'].notnull()]

# Calcular as mÃ©dias para ambos os casos
mean_missing = missing_customer_age.mean(numeric_only=True)
mean_not_missing = not_missing_customer_age.mean(numeric_only=True)

# Comparar as mÃ©dias
comparison = pd.DataFrame({
    "Mean_Customer_Age_Missing": mean_missing,
    "Mean_Customer_Age_Not_Missing": mean_not_missing
})

# Mostrar os resultados
comparison
```
